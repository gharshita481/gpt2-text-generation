```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

tokenizer.pad_token = tokenizer.eos_token
```

```python
def generate_text(input_text):
    inputs = tokenizer(
        input_text,
        return_tensors="pt",
        padding=True
    )

    output = model.generate(
        inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_length=80,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7
    )

    print("\nINPUT:", input_text)
    print("OUTPUT:", tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
generate_text("Once upon a time")
generate_text("The future of artificial intelligence is")
generate_text("India is a country where")
generate_text("Machine learning will change the world")
generate_text("Education is important because")
```
